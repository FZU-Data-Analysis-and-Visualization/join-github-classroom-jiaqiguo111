### 二、列与行操作

1. 选取 score 和 gender 两列。

   

2. 选取索引 0、2 的行。

   

3. 筛选 score>80 的行。

   

4. 新增一列 grade：score≥90 为优秀，否则为良好。

   

5. 删除 gender 列。

```python
import pandas as pd 
data_dict={'name':['Tom','Jerry','Mike'],'score':[85,92,78],'gender':['男','女','男']}
df1=pd.DataFrame(data_dict)

print('---df1前一行---')
print(df1.head(1))
print('---df1形状---')
print(df1.shape)
print('---df1列名---')
print(df1.columns.tolist())
print('---df1数据类型---')
print(df1.dtypes)

data_list=[[1,'A'],[2,'B'],[3,'C']]
df2=pd.DataFrame(data_list,columns=['id','tag'])
print('---df2---')
print(df2)

#------------------------二、列与行操作------------------------
print('-----1.选取 score 和 gender 两列')
print(df1[['score','gender']])
print('-----2.选取第0行和第2行')
print(df1.loc[[0,2]])
print('-----3.筛选 score 大于80的行')
print(df1[df1['score']>80])
df1['grade']=df1['score'].apply(lambda x:'优秀' if x>=90 else '良好')
print('-----4.添加 grade 列')
print(df1)
print('-----5.删除 grade 列')
df1=df1.drop(columns=['gender'])
print(df1)
```

### 三、缺失值

1. 创建含缺失值的 DataFrame：

   data = {"x":[1, None, 3], "y":[None, 5, 6]}

   检测缺失值并统计每列缺失数量。

   

2. 用 0 填充缺失值；删除含缺失值的行。

```python
import pandas as pd
data={'x':[1,None,3],'y':[None,5,6]}
df3=pd.DataFrame(data)
print('创建含缺失值的 DataFrame')
print(df3)
print('-----1.检测缺失值-----')
print(df3.isna())
print('-----2.每列缺失值数量-----')
print(df3.isna().sum())
df3_filled=df3.fillna(0)
print(df3_filled)
```

### 五、排序与去重

1. 按 score 升序排序。

   

2. 给 DataFrame 添加一行重复数据（如 Tom,85, 男），去重。

```python
import pandas as pd
data_dict={'name':['Tom','Jerry','Mike'],'score':[85,92,78],'gender':['男','女','男']}
df1=pd.DataFrame(data_dict)
print('-----1.按 gender 分组，score 的均值和最大值-----')
agg_score=df1.groupby('gender')['score'].agg(['mean','max'])
print(agg_score)

```
